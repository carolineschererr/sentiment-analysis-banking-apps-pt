```python
from google_play_scraper import app, Sort, reviews
import pandas as pd
```

```python
app_id = "com.itau" # trocar para o id de outros bancos para fazer scraping dos seus reviews

avaliacoes, _ = reviews(
    app_id,
    lang='pt',
    country='br',
    sort=Sort.NEWEST,
    count=50 # define o n√∫mero de reviews coletados
)
```

```python
avaliacoes
```

```python
df_avaliacoes = pd.DataFrame(avaliacoes)
```

```python
df_avaliacoes.score.value_counts()
```

```python
df_avaliacoes.info()
```

```python
df_avaliacoes = df_avaliacoes.drop_duplicates(subset=['reviewId'])
```

```python
df_avaliacoes.to_csv('avaliacoes_itau_total.csv', index=False, escapechar='\\')
```

```python
df_avaliacoes = df_avaliacoes[df_avaliacoes['score'] < 5]
```

```python
df_avaliacoes.to_csv('avaliacoes_itau_1_4.csv', index=False, escapechar='\\')
```
